```python
import yaml
from pathlib import Path
import numpy as np

from StarSharp import State, StateFactory
```

```python
directory = Path("/Users/jmeyers3/src/ts_config_mttcs/MTAOS/v13/ofc/")
senspath = directory / "sensitivity_matrix" / "lsst_sensitivity_dz_31_29_50.yaml"
with open(senspath, "r") as f:
    sens = np.array(yaml.safe_load(f))
normpath = directory / "normalization_weights" / "range-fwhm.yaml"
with open(normpath, "r") as f:
    norm = np.array(yaml.safe_load(f))
```

```python
factory = StateFactory(sens, norm=norm, use_dof="0-9,10-16,30-34", nkeep=12)

# Note, could also evaluate the DZ sens above anywhere in the field and use that
# as the input to the StateFactory.  The important bit is just that the length of
# the trailing index is the total number of degrees of freedom.
```

```python
state = factory.from_v([1]*12)
print("xstate: \n", state.x.value)
print()
print("fstate: \n", state.f.value)
print()
print("vstate: \n", state.v.value)
```

```python
state = factory.from_v(np.arange(12))
state2 = factory.from_x(state.x.value)
state3 = factory.from_f(state.f.value)

print(np.allclose(state2.f.value, state.f.value, atol=0, rtol=0))
print(np.allclose(state3.f.value, state.f.value, atol=0, rtol=0))
print()
print(np.allclose(state2.x.value, state.x.value, atol=0, rtol=0))
print(np.allclose(state3.x.value, state.x.value, atol=0, rtol=0))
print()
print(np.allclose(state2.v.value, state.v.value, atol=1e-14, rtol=0))
print(np.allclose(state3.v.value, state.v.value, atol=1e-14, rtol=0))
```
