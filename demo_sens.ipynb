```python
import yaml
from pathlib import Path
import numpy as np

from datatypes import State, StateFactory
```

```python
directory = Path("/Users/jmeyers3/src/ts_config_mttcs/MTAOS/v13/ofc/")
senspath = directory / "sensitivity_matrix" / "lsst_sensitivity_dz_31_29_50.yaml"
with open(senspath, "r") as f:
    sens = np.array(yaml.safe_load(f))
normpath = directory / "normalization_weights" / "range-fwhm.yaml"
with open(normpath, "r") as f:
    norm = np.array(yaml.safe_load(f))
```

```python
A = sens @ np.diag(norm)
factory = StateFactory(A, use_dof="0-9,10-16,30-35", nkeep=12)

# Note, could also evaluate the DZ sens above anywhere in the field and use that
# as the input to the StateFactory.  The important bit is just that the length of
# the trailing index is the total number of degrees of freedom.
```

```python
state = factory.from_v([1]*12)
print("xstate: \n", state.x.state)
print()
print("fstate: \n", state.f.state)
print()
print("vstate: \n", state.v.state)
```

```python
state = factory.from_v(np.arange(12))
state2 = factory.from_x(state.x.state)
state3 = factory.from_f(state.f.state)

print(np.allclose(state2.f.state, state.f.state, atol=0, rtol=0))
print(np.allclose(state3.f.state, state.f.state, atol=0, rtol=0))
print()
print(np.allclose(state2.x.state, state.x.state, atol=0, rtol=0))
print(np.allclose(state3.x.state, state.x.state, atol=0, rtol=0))
print()
print(np.allclose(state2.v.state, state.v.state, atol=1e-14, rtol=0))
print(np.allclose(state3.v.state, state.v.state, atol=1e-14, rtol=0))
```
